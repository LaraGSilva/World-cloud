```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Importando o relatório de vendas

import pandas as pd

# Caminho para o arquivo
file_path = '/Users/laragoncalves/Downloads/LARA.txt'

# Tentando ler o arquivo com a codificação ISO-8859-1
try:
    entrevista_lara = pd.read_csv(file_path, delimiter='\t', encoding='ISO-8859-1')
    display(entrevista_lara)
except UnicodeDecodeError as e:
    print(f"Erro de codificação: {e}")
except FileNotFoundError as e:
    print(f"Arquivo não encontrado: {e}")

#entrevista_lara.head(5)
#print(type(entrevista_lara))

import re
from collections import Counter


texto_inteiro = ''.join(entrevista_lara["FALAS - LARA"].astype(str))
```

```python
palavras = re.findall(r'\b\w+\b', texto_inteiro.lower())
print(palavras)
```

```python

palavras_nao_incluidas = ['eu','tem', 'com','isso', 'pessoa','o','no','do','por','ser','até','porque','por que','da','também','esse','vou','muito','a','um','e','mais','então','ou','você','na','me','enfim','ah','que', 'acho', 'para', 'de', 'essa', 'mas', 'mesmo', 'em', 'é', 'está', 'uma', 'não']

palavras_filtradas = [
        
        palavra for palavra in palavras if palavra not in palavras_nao_incluidas]

print(palavras_filtradas)
```

```python
text = Counter(palavras_filtradas)

#print(text)

palavras_ordenadas = sorted(text.items(), key=lambda x: x[1], reverse=True)

for palavra, count in palavras_ordenadas:
    
        print(f"{palavra}: {count}")

```

```python
texto_filtrado = ' '.join(palavras_filtradas)
type(palavras_ordenadas)

image_mask = ~np.array(Image.open('mapa_br.png'))

def create_cloud(palavras_usadas):
    wordcloud = WordCloud(
        width=image_mask.shape[1], 
        height=image_mask.shape[0],
        mask= image_mask, 
        colormap = 'Dark2_r',
        background_color='black').generate(palavras_usadas)


    plt.figure(figsize=(15, 10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

create_cloud(texto_filtrado)
```
